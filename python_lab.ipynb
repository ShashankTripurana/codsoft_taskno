```python
import numpy as np
from scipy.signal import square,sawtooth
import matplotlib.pyplot as plt
time_vector=np.linspace(0,10,1000)
square_wave=square(time_vector,0.5)
sawtooth_wave=sawtooth(time_vector,0.5)
plt.figure(figsize=(10,12))
plt.subplot(1,2,1)
plt.plot(time_vector,square_wave)
plt.grid(True)
plt.subplot(1,2,2)
plt.plot(time_vector,sawtooth_wave)
plt.grid(True)
plt.show()
```
